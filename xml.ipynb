### Elements contain text

```python
root = etree.Element('root')
root.text = 'TEXT'
print(root.text)
etree.tostring(root)
```

```python
html = etree.Element('html')
body = etree.SubElement(html, 'body')
body.text = 'TEXT'
etree.tostring(html, pretty_print=True)
```

```python
br = etree.SubElement(body, 'br')
etree.tostring(html)
```

```python
br.tail = 'TAIL'
etree.tostring(html)
```

```python
etree.tostring(br)
```

```python
etree.tostring(br, with_tail=False)
```

```python
etree.tostring(html, method='text')
```

# Basic Usage of lxml
#### https://python.keicode.com/advanced/xml-lxml-1.php

```python
tree = etree.parse('usa.xml')
root = tree.getroot()
print('root tag = %s' % root.tag)
print(etree.tostring(tree).decode())
```

```python
print( '== state ==')
for state in root.iter('state'):
    print('*** %s' % state.attrib)
    print('attr=[%s]' % state.attrib['code'])
    for child in state:
        print(' tag=[%s] text=[%s]' % (child.tag, child.text))
```

```python
states = tree.xpath('/states/state')
if len(states):
    for state in states:
        print( state.attrib)
```

```python
type(states)
```

```python
states = tree.xpath('//state')
if len(states):
    for state in states:
        print(state.attrib)
```

```python
states = tree.xpath('//state[@code="HI"]')
if len(states):
    for state in states:
        print(state.attrib)
        print(state.text)
        print(state.tag)
```
