```python
tree = et.ElementTree(file='sports.xml')
root = tree.getroot()
```

```python
for fruits in root:
    for fruit in fruits:
        for item in fruit:
            print(item.tag, ':', item.text)
```

```python
url = "https://www.uniprot.org/uniprot/" + keyword + ".xml"
f = urlopen(url)
xml_data = f.read()
tree = et.ElementTree(xml_data)
root = tree.getroot()
```

```python
from xml.etree import ElementTree
tree = ElementTree.parse('sports.xml')

# 先頭要素を表す Element オブジェクトを取得
elem = tree.getroot()
print(elem.tag)     #=> tree
print(elem.attrib)  #=> {'name': 'hello'}
```

```python
elem = ElementTree.fromstring('<tree name="hello"><trunk>...</trunk></tree>')
print(elem.tag)     #=> tree
print(elem.attrib)  #=> {'name': 'hello'}
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse('sports.xml')
root = tree.getroot()
```
